```python
my_list = [1, 2, 3] 

### multiply by two
my_list_mult = [1, 2, 3] * 2
my_list_mult   # -> [1, 2, 3, 1, 2, 3]
## instead one can do
a = 2
my_list_mult1 = [a * i for i in my_list] 
my_list_mult1   # -> [2, 4, 6]
## in numpy 
my_array = np.array([1, 2, 3])
my_array_mult = my_array * 2
my_array_mult   # -> [2, 4, 6]
```
